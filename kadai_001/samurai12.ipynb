```python
from datetime import date
from openpyxl import Workbook

# 今日の日付
today = date.today()
today_cell_format = "yyyy/mm/dd"   # セル表示用
today_str_for_filename = today.strftime("%Y%m%d")  # ファイル名用（/は使えないため）

# Excelブック作成
wb = Workbook()
ws = wb.active
ws.title = "請求書"

# タイトル・会社情報
ws["B2"] = "請求書"
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

# No. と日付
ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = today
ws["G5"].number_format = today_cell_format  # YYYY/MM/DD 表示

# 明細ヘッダー
ws["B10"] = "商品名"
ws["C10"] = "数量"
ws["D10"] = "単価"
ws["E10"] = "金額"

# 明細
ws["B11"] = "商品A"
ws["C11"] = 2
ws["D11"] = 10000
ws["E11"] = "=C11*D11"

ws["B12"] = "商品B"
ws["C12"] = 1
ws["D12"] = 15000
ws["E12"] = "=C12*D12"

# 小計
ws["E13"] = "=E11+E12"

# 合計
ws["B15"] = "合計"
ws["E15"] = "=E13"

# 消費税（10%）
ws["B16"] = "消費税"
ws["E16"] = "=E15*10%"

# 税込合計
ws["B17"] = "税込合計"
ws["E17"] = "=E15+E16"

# 保存（ファイル名）
filename = f"請求書_{today_str_for_filename}.xlsx"
wb.save(filename)
print(f"Created: {filename}")
```

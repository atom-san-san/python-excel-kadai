```python
import pandas as pd

data = [
    {"日付": "2023-05-17", "社員名": "山田", "売上": 100, "部門": "メーカー"},
    {"日付": "2023-05-18", "社員名": "佐藤", "売上": 200, "部門": "代理店"},
    {"日付": "2023-05-19", "社員名": "鈴木", "売上": 150, "部門": "メーカー"},
    {"日付": "2023-05-20", "社員名": "田中", "売上": 300, "部門": "商社"},
    {"日付": "2023-05-21", "社員名": "高橋", "売上": 250, "部門": "代理店"},
]

df = pd.DataFrame(data)

# ★ 日付から時間を削除
df["日付"] = pd.to_datetime(df["日付"]).dt.date

avg_sales = df["売上"].mean()
df["平均売上"] = avg_sales

def performance(sales, avg):
    if sales >= avg + 50:
        return "A"
    elif sales >= avg:
        return "B"
    else:
        return "C"

df["業績ランク"] = df["売上"].apply(lambda x: performance(x, avg_sales))

df.to_excel("業績.xlsx", index=False)
```
